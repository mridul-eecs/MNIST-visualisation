```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../digit-recognizer/train.csv')
df.head(5)
```

```python
labels = df['label']
labels[:5]
```

```python
features = df.drop('label', axis=1)
features.head(5)
```

```python
# Standard normalisation:
from sklearn.preprocessing import StandardScaler
features = StandardScaler().fit_transform(features)
```

```python
# get eigen values and eigenvectors

# covariance matrix
covar = np.matmul(features.T, features)
print(covar.shape)
from scipy.linalg import eigh
values, vectors = eigh(covar, eigvals=(782,783))
```

```python
new_coordinates = np.matmul(vectors.T, features.T)
```
